```python
import numpy as np
```

```python
samples = np.random.rayleigh(scale=6.5,size=(500,40))
```

```python
mvue_values = np.sqrt(np.mean(samples**2,axis=1)/2)
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
import scipy.stats as s
```

```python
sns.distplot(a=mvue_values,bins=8,kde=False,fit=s.norm)

plt.xlabel("Mvue values of population parameter of Rayleigh Distribution")

plt.ylabel("P(MVUE = k)")
```

```python
unbiased_estimate = np.mean(mvue_values)
```

```python
unbiased_estimate
```

```python
np.std(mvue_values)*1.96
```

```python
samples_1 = np.random.normal(loc=180,scale=5,size=(500,40))
```

```python
samples_2 = np.random.normal(loc=170,scale=10,size=(500,40))
```

```python
sample_means_1 = np.mean(samples_1,axis=1)
```

```python
sample_means_2 = np.mean(samples_2,axis=1)
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
sns.distplot(a=sample_means_1,bins=8,kde=False,fit=s.norm)

plt.xlabel("X_bar_1")

plt.ylabel("Probability (X_bar_1 = k)")
```

```python
best_estimate_1 = np.mean(sample_means_1)
```

```python
best_estimate_1
```

```python
np.std(sample_means_1)
```

```python
5/np.sqrt(40)
```

```python
sns.distplot(a=sample_means_2,bins=8,kde=False,fit=s.norm)

plt.xlabel("X_bar_2")

plt.ylabel("Probability(X_bar_2 = k)")
```

```python
best_estimate_2 = np.mean(sample_means_2)
```

```python
best_estimate_2
```

```python
np.std(sample_means_2)
```

```python
df = pd.DataFrame(data={'X_bar_1':sample_means_1,'X_bar_2':sample_means_2,'X_bar_1 - X_bar_2':(sample_means_1-sample_means_2)})
```

```python
df
```

```python
sns.distplot(a=df['X_bar_1 - X_bar_2'],bins=8,kde=False,fit=s.norm)

plt.xlabel("X_bar_1 - X_bar_2")

plt.ylabel("Probability (X_bar_1 - X_bar_2 = k)")
```

```python
best_estimate = np.mean(df['X_bar_1 - X_bar_2'])
```

```python
best_estimate
```
